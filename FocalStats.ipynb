```python
# First loop delveloped. Show logic of a moving window and determine how to access each desired element. 
i = 0
j = 0
while i < 3:
    while j < 3:
        print(array[i][j])
        j += 1
    i += 1
    j = 0
```

```python
# Modify first loop to determine how to 'move' window horizontally. 
i = 0
j = 0
iteration = 0
while iteration <= 2:
    while i < 3:
        while j > iteration and j < 3:
            print(array[i][j])
            j += 1
        i += 1
        j = 0
    iteration += 1
```
